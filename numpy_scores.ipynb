```python
row_min = np.min(row)
row_max = np.max(row)
print(row_min,row_max)
```

```python
Normalize = (scores - row_min) / (row_max - row_min)
```

```python
print(Normalize)
```

```python
max_index = np.argmax(Normalize)
print(max_index)
```

```python
row, col = np.unravel_index(max_index, Normalize.shape)
print("Student index (row):", row)
print("Subject index (column):", col)
```
